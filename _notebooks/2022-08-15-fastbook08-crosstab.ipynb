```python
from fastai.collab import *
from fastai.tabular.all import *
```

```python
path = untar_data(URLs.ML_100k)
```

```python
df = pd.read_csv(path/'u.data', delimiter='\t', header=None,
                names=['user', 'movie', 'rating', 'timestamp'])
df = df.drop(columns='timestamp')
df
```

```python
def f(movie_id):
    x = df[df.movie==movie_id].drop(columns='movie')
    x.columns = ['user', movie_id]
    return x

xs = [f(o) for o in df.movie.value_counts().index[:9]]
```

```python
x, xs = xs[0], xs[1:]
for o in xs:
    x = x.merge(o, how='outer')
```
